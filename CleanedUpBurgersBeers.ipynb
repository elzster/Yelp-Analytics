## Let's get a burger and beer

```python
import json
import pandas as pd
import numpy as np
import requests
from config import api_key
```

```python
## define a function to search for restaurants that have burgers or beer 
```

```python
name = []
city = []
state = []
category = []
latitude = []
longitude = []
price = []
review_count = []
business_id = []
rating = []
```

```python
def search_soul(set_num): 
# This function finds burgers and beer in New Brunswick NJ

    url = 'https://api.yelp.com/v3/businesses/search'
    headers = {
        'Authorization': 'Bearer {}'.format(api_key),
    }
    url_params = { #parameters passed to the API
    "categories": 'burger, All' or 'Beer, All' or 'pub, All' or 'burgers, All' or 'tavern, All',
    "location":'New Brunswick, NJ',
#     "alias": "bar" or "Bars" or "gastropub" or "tavern" or "beers"
    'offset': offset_num, # We are going to iterate the offset
     "limit":50 # Maximum return of results per request per yelp. we'll take 19 of those yelp. with cheese and bacon.
     }

    response = requests.get(url, headers=headers, params=url_params)
    return response.json() # Returns a JSON.
```

## Get a Flight 

```python
if __name__ == "__main__":
    for offset_num in np.arange(50,550,50) : 
# get 500 results, in steps of 50 results per request.

        try:
            output_json = search_soul(offset_num) # flipping burgers and drinking beer
            businesses=output_json['businesses']
            for business in businesses:
                name.append(business['name'])
                business_id.append(business['id'])
                city.append(business['location']['city'])
                state.append(business['location']['state'])
                latitude.append(business['coordinates']['latitude'])
                longitude.append(business['coordinates']['longitude'])
                review_count.append(business['review_count'])
                rating.append(business['rating'])
                category.append(business['categories'][0]['title'])

                try:
                    price.append(business['price'])
                except KeyError:
                    price.append(0)
            print(offset_num) # Making sure each offset iteration is running
            print(output_json) # If you wanna check the JSON for each iteration
            if offset_num == 50:
                df_first = pd.DataFrame.from_dict(output_json['businesses'])

# gimmie an 'a': Ey. Where's my beer? ... documentation said to use primary key 'businesses'
# gimmie a 'p!': parsing the JSON. 
            else:
                df2 = pd.DataFrame.from_dict(output_json['businesses'])
                df_first = df_first.append(df2)
# The conditional statement above is so that I can append my results into a single dataframe, to 
# save into a single csv document.
        except AttributeError:
            print("error at ", offset_num) # Helpful for debugging purposes
            
#     df_first.to_csv("output_burgers_and_beer_data.csv", index = False)
            
```

```python
business_df = pd.DataFrame({
    "Business" : name,
    "Business ID": business_id,
    "City": city,
    "State": state,
    "Category": category,
    "Latitude": latitude,
    "Longitude": longitude,
    "Price":price,
    "Reviews": review_count
})
# len(price)
```
