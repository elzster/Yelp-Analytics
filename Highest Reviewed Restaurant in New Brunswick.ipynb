```python
new_brunswick_best_burger = pd.read_csv("data/New Brunswick, NJ_Burger.csv")
new_brunswick_best_burger2 = new_brunswick_best_burger.sort_values(by='Reviews', ascending=False).head(30)
```

```python
master = new_brunswick_best_burger2.drop(columns=['Unnamed: 0', 'Business ID'])
master.head(25)
```

```python
business_name = master[["Business"]]
business = master[["Business", "Reviews"]]
coordinates = master[["Latitude", "Longitude"]]

reviews = master[["Reviews"]].astype(float)
# coordinates
```

```python
fig = gmaps.figure()

markers = gmaps.marker_layer(coordinates)

# fig.add_layer(markers)

heatmap_layer = gmaps.heatmap_layer(coordinates)
heatmap_layer.max_intensity = 45
heatmap_layer.point_radius = 75

fig.add_layer(heatmap_layer)

fig
```

```python
x_axis = business["Business"]

plt.figure(figsize=(40,40))
plt.bar(business["Business"], business["Reviews"], color="orange", alpha=.5)

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis, rotation=90, weight="bold", fontsize=25);

plt.title("Highest Reviewed Restaurant in New Brunswick\n\n", weight="bold", fontsize=35);
plt.ylabel("Reviews\n\n", weight="bold", fontsize=35);
plt.xlabel("\n\n\nRestaurants\n\n\n", weight="bold", fontsize=35);
```
